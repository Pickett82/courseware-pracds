## Problems with Scoring

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
clf = KNeighborsClassifier(1).fit(tips[['total_bill', 'tip']], tips['sex'])
```

```python
clf.score(tips[['total_bill', 'tip']], tips['sex'])
```

* this is great at predicting "remembering" data already seen
* very very probably terrible at any new data

```python
from sklearn.model_selection import train_test_split
```

```python
Xtr, Xte, ytr, yte = train_test_split(tips[['total_bill', 'tip']], tips['sex'])
clf = KNeighborsClassifier(1).fit(Xtr, ytr)
clf.score(Xte, yte)
```
