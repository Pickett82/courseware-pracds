## How do I use pandas to plot?

Pandas has a `.plot` which you configure using its arguments (eg., `kind=hist`), this however *just uses* seaborn. 

```python
import pandas as pd
ti = pd.read_csv('datasets/titanic.csv')
ti.head(3)
```

```python
ti['age'].plot(kind='hist')
plt.title("Histogram of Titanioc Passenger Ages")
```

Often pandas doesn't choose the right plot, or the right dataseries... therefore it's often easier to go straight to using seaborn. 

## How does Matplotlib make graphs

Eaxh matplotlib graph consists of:
* A `figure` object (the "canvas")
* An `axes` (or multiple axes) object(s)
* Glyphs/artists

generate canvas -> put axes on canvas -> put glyphs on the axes

## Object Oriented API

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(15, 8)) # create figure
ax = fig.add_subplot(111)         # create axes

ax.scatter(ti['fare'], ti['age']) # add data to axes
ax.scatter(ti.groupby('pclass').mean()['fare'], ti.groupby('pclass').mean()['age'])

ax.set_xlabel('fare')
ax.set_ylabel('age')
ax.set_title("Titanic Passenger Age vs. Fare")
```

```python
ax.get_xlabel()
```

## Matlab-like API

```python
fig, axes = plt.subplots(2, 1, figsize=(15, 8))

axes[0].scatter(ti['fare'], ti['age'], )
axes[1].scatter(ti.groupby('pclass').mean()['fare'], ti.groupby('pclass').mean()['age'])

axes[0].set_xlabel('fare')
axes[0].set_ylabel('age')
axes[0].set_title("Titanic Passenger Age vs. Fare")
#plt.savefig("my_titanic_graphic.png")
```

## How do I use seaborn to visualize data?

* supply x (, y, etc.) as individual arguments

## How do I create a distribution plot?

`kde` will infer a continuous frequency line, set `False` for just bars:

```python
fig = plt.figure(figsize=(15, 8))
sns.distplot(ti['age'], vertical=False, kde=True, rug=True);
```

## How do I create a violin plot?

The width of a violin plot is the frequency (ie., how common) a value is within some column... 

```python
x = 1; y = 2; z = 10
z;
```

```python
plt.style.available
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(15, 8))
sns.violinplot(data=ti, x='age', y='class', order=['First', 'Second', 'Third']);
```

A violinplot uses *area* to visualize frequency, rather than height as in a barplot. 

## How do I create a box plot?

Box plots show quarties (25th, 50th, 75th) and outliers. 

```python
fig = plt.figure(figsize=(15, 8))
sns.boxplot(data=ti, y='age', x='embark_town');
```

## What are the core multivariate plots?

* columns factored by a categorical column
    * sns.violinplot
    * sns.barplot

* real-valued columns correlated with each other: 
    * sns.lineplot
    * sns.scatterplot

## How do I create a violin plot for multiple columns?

A violin plot can show multiple distributions, each a subset of a single column, factored (or grouped) by another.

```python
sns.violinplot(data=ti, x='survived', y='age');
```

...plotting the age distribution, split by `survived`.

```python
sns.violinplot(data=ti, x="age", y='embark_town');
```

## How do I create a bar plor for multiple columns?

Bar plots are useful for discrete data or for showing, here, the means:

```python
fig = plt.figure(figsize=(15, 8))
sns.barplot(data=ti, x='survived', y='age');
```

## How do I create a line plot?
